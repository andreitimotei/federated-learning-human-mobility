```python
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics
df = pd.read_csv("federated_training_metrics.csv")

# Plot Duration Metrics
plt.figure(figsize=(10, 5))
plt.plot(df["round"], df["duration_loss"], label="Duration Loss")
plt.plot(df["round"], df["duration_mae"], label="Duration MAE")
plt.title("Duration Prediction Metrics Over Rounds")
plt.xlabel("Round")
plt.ylabel("Loss / MAE")
plt.legend()
plt.grid(True)
plt.show()

# Plot Destination Metrics
plt.figure(figsize=(10, 5))
plt.plot(df["round"], df["destination_loss"], label="Destination Loss")
plt.plot(df["round"], df["destination_accuracy"], label="Destination Accuracy")
plt.title("Destination Prediction Metrics Over Rounds")
plt.xlabel("Round")
plt.ylabel("Loss / Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
